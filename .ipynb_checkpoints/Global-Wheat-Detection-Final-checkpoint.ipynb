````python
# Grabbing all the unique image id's from the DataFrame.

image_ids = df_train["image_id"].unique()
print("Few Unique ids:",image_ids[:3],"\n\nNumber of unique images:", len(image_ids))
````

````python
# Splitting to Train and validation datasets from the training dataset.

train_image_ids = image_ids[0:3363]
val_image_ids = image_ids[3363:3373]
````

````python
# Grouping/Mapping the bounding boxes to their respective Image_ids

def group_bboxes(row):
    bounds = row['bbox'].str.split(",", expand=True)
    bounds[0] = bounds[0].str.slice(start=1)
    bounds[3] = bounds[3].str.slice(stop=-1)
    
    return bounds.values.astype(float)

bboxes = df_train.groupby('image_id').apply(group_bboxes)
print("Image_id: {}".format(train_image_ids[1]), "\n\nFew bounding boxes in the image\n",bboxes[train_image_ids[1]][:5])
````

````python
# Reads the Image from the disk, resizes the image to (256, 256). 

def load_images(image_id):
    '''
    This function reads the image from the disk and returns the image resized to (256, 256)
    
    Arguments:
        image_id: The image id of the image that should be read from the disk.
        
    Returns:
        
        An image resized to (256, 256)
        return type: numpy.ndarray
        
    '''
    
    load_images_path = "./dataset/train/" # Change the path to `../input/global-wheat-detection/train/` for working on Kaggle
    
    global use_GPU
    if use_GPU:
        image = cv2.UMat(cv2.imread(load_images_path+image_id+'.jpg'))
    else:
        image = cv2.imread(load_images_path+image_id+'.jpg')
        
    image = cv2.resize(image, (256, 256))
    
    if use_GPU:
        return cv2.UMat.get(image)
    else:
        return image
                 
                 
````

````python
# Creating the training dictionaries containing images and bounding boxes with image_ids as keys.
# Bounding boxes are scaled by 4, as original image of size (1024, 1024) is resized to (256, 256) i.e scaled by 1024/256 = 4

train_images = {}
train_bboxes = {}
for image_id in tqdm(train_image_ids):
    train_images[image_id] = load_images(image_id)
    train_bboxes[image_id] = bboxes[image_id].copy()/4
````

````python
# Creating the validation dictionaries containing images and bounding boxes with image_ids as keys.
# Bounding boxes are scaled by 4, as original image of size (1024, 1024) is resized to (256, 256) i.e scaled by 1024/256 = 4

val_images = {}
val_bboxes = {}
for image_id in tqdm(val_image_ids):
    val_images[image_id] = load_images(image_id)
    val_bboxes[image_id] = bboxes[image_id].copy()/4
````

````python
def draw_bbox(image, bbox):
    
    '''
    Returns an image with a single bounding box drawn.
    
    Arguments:
        image: input image
        
        bbox: a single bounding box in the 
        
    Returns:
        
        image with bounding box drawn.
        return type: numpy.ndarray.
    
    '''
    
    x, y, w, h = bbox
    image = cv2.rectangle(image,
                         (int(x), int(y)), 
                         (int(x+w), int(y+h)),
                         (255, 0, 0),
                         1)
    return image

def draw_bound_boxes(image_id, bboxes, source="train"):
    
    '''
    This function reads the image corresponding to the image_id from disk.
     It draws multiple bounding boxes on the given image using draw_bbox function.
    
    Arguments:
        image_id: The image id of the image that should be read from the disk.
        
        bboxes: A list of bounding boxes pertaining to the image_id
        
        source (optional): The name of the folderfrom which the images should be read.
        
    Returns: 
        An image with all of it bounding boxes drawn over it.
        return_type: numpy.ndarray
    '''
    
    bboxes_path = './dataset/'# Change the path to `../input/global-wheat-detection/train/` for working on Kaggle
    if use_GPU:
        image = cv2.UMat(cv2.imread(bboxes_path+source+'/'+image_id+'.jpg'))
    else:
        image = cv2.imread(bboxes_path+source+'/'+image_id+'.jpg')
        
    image = cv2.resize(image, (256, 256))
        
    for bbox in bboxes:
        image = draw_bbox(image, bbox)
    if use_GPU:
        return cv2.UMat.get(image)
    else:
        return image
````

````python
# We could check for leading bounding boxes which are out of the image, but not implimented due to time constraint.

def show_images(image_ids, bboxes, source='train'):
    
    '''
    It plots the images with bounding boxes drawn.
    We could check for leading bounding boxes which are out of the image, \
    but not implimented due to time constraint.

    Arguments: 
        
        image_ids: A list of image_ids which are to be plotted
        
        bboxes: A dictionary containing the bounding boxes mapped to their respective image_ids as keys.
        
        source (optional): The name of the folderfrom which the images should be read.
        
    Returns:
        This function plots the given images but returns None.
    
    '''
    
    images = []
    global use_GPU
    for image_id in image_ids:
        images.append(
                draw_bound_boxes(image_id, bboxes[image_id], source)
            )
    num_of_images = len(image_ids)
    fig, axes = plt.subplots(1, num_of_images, figsize=(10*num_of_images, 10*num_of_images))
    
    for i, image in enumerate(images):
        if use_GPU:
            axes[i].imshow(cv2.UMat.get(image))
        else:
            axes[i].imshow(image)
````

````python
def draw_grid_lines(image_id, labels, grid_line_distance=32, source='train'):
    global use_GPU
    
    image_path = '../input/global-wheat-detection/train/'+source+'/'+image_id+'.jpg'
    if use_GPU:
        image = cv2.UMat(draw_bound_boxes(image_id, labels))
    else:
        image = draw_bound_boxes(image_id, labels)
    image = cv2.resize(image, (256, 256))
    
    # Vertical Lines
    prev_x = 0
    prev_y = 0
    for dist in range(0, 256, grid_line_distance):
        image = cv2.line(image, (prev_x, prev_y+(dist-prev_y)), (prev_x+256, prev_y+(dist-prev_y)), (0, 0, 255), 1, 1)
        prev_x = prev_x
        prev_y = dist
    
    # Horizontal Lines
    prev_x = 0
    prev_y = 0
    for dist in range(0, 256, grid_line_distance):
        image = cv2.line(image, (prev_x+(dist-prev_x), prev_y), (prev_x+(dist-prev_x), prev_y+256), (0, 0, 255), 1, 1)
        prev_x = dist
        prev_y = prev_y
        
    if use_GPU:
        image = cv2.UMat.get(image)
        
    else:
        image = image

    return image
````

* The custom-model is similar to a Yolo-v2 model (with added Residual connections) with two anchor points per grid.
* According to the model, each grid can detects two bounding boxes.
* Smaller the grid size, more are the grids, hence more bounding boxes are detected per image with increased training time.
* Therefore selecting `8*8` grid size seems optimal.

# Data Generator

```python
The label output of the Data Generator has a shape of [BATCH_SIZE, 32, 32, 10]
```

The 10 outputs of each grid are formatted as below
- [0]: Confidence probability of anchor one

- [1,2]: x,y-center of anchor one

- [3,4]: w,h of anchor one

- [5]: Confidence probability of anchor two

- [6,7]: x,y-center of anchor two

- [8,9]: w,h-center of anchor two

````python
class DataGenerator(tf.keras.utils.Sequence):
    
    '''
    Data Generator class which is a child class of tf.keras.utils.Sequence
    
    Arguments:
        image_ids: The names of the images in the data folder removing their extensions.
        
        image_pixels: A dictionary containing the images in nummpy.ndarrays format, mapped to the respective image_ids. \
                When called with image id, the dictionary should return an image with the shape of (256, 256, 3) corresponding to the respective image ID.
        
        labels: A dictionary containing the cleaned and adjusted  bounding-boxes which fit the image shape of (256, 256).\
                When called with the image_id, the dictionary should return all the bounding boxes pertaining to the respective image in the list type.
        
        batch_size: The batch size being used for the data generation.
        
        shuffle (optional): A bool type input, Should be `True` if the dataset should be shuffled after every epoch, else should be set to `False`. 
                Default Value: False
        
        augment (optional): A bool type argument, Should be `True` uf the images should be augmented before being fed to the model. 
                Default value: False
                
    
    
    '''

    def __init__(self, image_ids, image_pixels: dict, labels: dict, batch_size: int=1, shuffle: bool=False, augment: bool=False):
        self.image_ids = image_ids
        self.image_pixels = image_pixels
        self.labels = labels
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.augment = augment
        
        self.on_epoch_end()
        self.image_grid = self.form_image_grid()
        
    
    def form_image_grid(self):    
        
        # Forming the image grid for the retriving the bounding-boxes post prediction.
        
        image_grid = np.zeros((32, 32, 4))

        # x, y, width, height
        cell = [0, 0, 256 / 32, 256 / 32] 

        for i in range(0, 32):
            for j in range(0, 32):
                image_grid[i,j] = cell

                cell[0] = cell[0] + cell[2]

            cell[0] = 0
            cell[1] = cell[1] + cell[3]

        return image_grid
    
    def __len__(self):
        
        '''
        Returns the Length of the training data, Needed by tensorflow Data Generator, which is called internally during Training of the model.
        '''
        
        return int(np.floor(len(self.image_ids) / self.batch_size))
    
    
    def on_epoch_end(self):
        
        """
        Shuffles the dataset after every Epoch to reduce overfitting during training.
        """
        
        self.indexes = np.arange(len(self.image_ids))
        
        if self.shuffle == True:
            np.random.shuffle(self.indexes)
    
    
    def __data_generation(self, batch_ids):
        
        # A Function which is a Wrapper for all the data augmentation of images and bounding-boxes
        # Returns a list of 
        
        X, y = [], []

        # Generate data
        for i, image_id in enumerate(batch_ids):
            pixels = self.image_pixels[image_id]
            bboxes = self.labels[image_id]

            if self.augment:     
                pixels, bboxes = self.augment_image(pixels, bboxes)
            else:
                pixels = self.contrast_image(pixels)
                bboxes = self.form_label_grid(bboxes)

            X.append(pixels)
            y.append(bboxes)

        return np.array(X), np.array(y)

            
    def __getitem__(self, index):
        
        # Return the images and bounding boxes with augmentations applied
        # This is internally called by tensorflow during training.
        
        indexes = self.indexes[index * self.batch_size:(index+1) * self.batch_size]
        
        batch_ids = [self.image_ids[i] for i in indexes]
        X, y = self.__data_generation(batch_ids)
        
        return X, y
        
    def augment_image(self, pixels, bboxes):
        
        # Applies augmentation to the images and returns the images
        
        bbox_labels = np.ones(len(bboxes))

        aug_result = self.train_augmentations(image=pixels, bboxes=bboxes, labels=bbox_labels)

        bboxes = self.form_label_grid(aug_result['bboxes'])

        return np.array(aug_result['image']) / 255, bboxes
    
    def contrast_image(self, pixels):        
        aug_result = self.val_augmentations(image=pixels)
        return np.array(aug_result['image']) / 255
    
    def form_label_grid(self, bboxes):
        
        '''
        This function returns the image grid for labels (Bounding boxes).
        '''
        
        label_grid = np.zeros((32, 32, 10))

        for i in range(0, 32):
            for j in range(0, 32):
                cell = self.image_grid[i,j]
                label_grid[i,j] = self.rect_intersect(cell, bboxes)

        return label_grid
    

    def rect_intersect(self, cell, bboxes):
        
        '''
        Sets the bounding boxes to the anchor points for each grid in the image.
        The return data format is [BATCH_SIZE, 32, 32, 10], this is the shape of the output layer.
        
        The 10 outputs of each grid are formatted as below
        - [0]: Confidence probability of anchor one

        - [1,2]: x,y-center of anchor one

        - [3,4]: w,h of anchor one

        - [5]: Confidence probability of anchor two

        - [6,7]: x,y-center of anchor two

        - [8,9]: w,h-center of anchor two
        '''
        
        cell_x, cell_y, cell_width, cell_height = cell
        cell_x_max = cell_x + cell_width 
        cell_y_max = cell_y + cell_height

        anchor_one = np.array([0, 0, 0, 0, 0])
        anchor_two = np.array([0, 0, 0, 0, 0])

        # check all boxes
        for bbox in bboxes:
            box_x, box_y, box_width, box_height = bbox
            box_x_centre = box_x + (box_width / 2)
            box_y_centre = box_y + (box_height / 2)

            if(box_x_centre >= cell_x and box_x_centre < cell_x_max and box_y_centre >= cell_y and box_y_centre < cell_y_max):

                if anchor_one[0] == 0:
                    anchor_one = self.yolo_shape(
                        [box_x, box_y, box_width, box_height], 
                        [cell_x, cell_y, cell_width, cell_height]
                    )

                elif anchor_two[0] == 0:
                    anchor_two = self.yolo_shape(
                        [box_x, box_y, box_width, box_height], 
                        [cell_x, cell_y, cell_width, cell_height]
                    )

                else:
                    break

        return np.concatenate((anchor_one, anchor_two), axis=None)
    
    def yolo_shape(self, box, cell):
        
        '''
        Returns the bounding boxes adjusted following the yolo format.
        '''
        
        box_x, box_y, box_width, box_height = box
        cell_x, cell_y, cell_width, cell_height = cell

        # top left x,y to centre x,y
        box_x = box_x + (box_width / 2)
        box_y = box_y + (box_height / 2)

        # offset bbox x,y to cell x,y
        box_x = (box_x - cell_x) / cell_width
        box_y = (box_y - cell_y) / cell_height

        # bbox width,height relative to cell width,height
        box_width = box_width / 256
        box_height = box_height / 256

        return [1, box_x, box_y, box_width, box_height]
    
    
````

````python
DataGenerator.train_augmentations = albu.Compose([
        albu.RandomSizedCrop(
            min_max_height=(200, 200), 
            height=256, 
            width=256, 
            p=0.8
        ),
        albu.OneOf([
            albu.Flip(),
            albu.RandomRotate90(),
        ], p=1),
        albu.OneOf([
            albu.HueSaturationValue(),
            albu.RandomBrightnessContrast()
        ], p=1),
        albu.OneOf([
            albu.GaussNoise(),
            albu.GlassBlur(),
            albu.ISONoise(),
            albu.MultiplicativeNoise(),
        ], p=0.5),
        albu.Cutout(
            num_holes=8, 
            max_h_size=16, 
            max_w_size=16, 
            fill_value=0, 
            p=0.5
        ),
        albu.CLAHE(p=1),
        albu.ToGray(p=1),
    ], 
    bbox_params={'format': 'coco', 'label_fields': ['labels']})

DataGenerator.val_augmentations = albu.Compose([
    albu.CLAHE(p=1),
    albu.ToGray(p=1),
])
````

````python
BATCH_SIZE = 2

train_generator = DataGenerator(
    train_image_ids,
    train_images,
    train_bboxes, 
    batch_size=BATCH_SIZE, 
    shuffle=True,
    augment=True
)

val_generator = DataGenerator(
    val_image_ids, 
    val_images,
    val_bboxes, 
    batch_size=10,
    shuffle=False,
    augment=False
)

image_grid = train_generator.image_grid
````

# Model Creation Functions
* This model is a combination of concepts from Yolo and Resnet.
* The model has two paths mainly "main path" and "residual/shortcut path"
* Each Yolo block is modified to accept a Residual path and this block is named "Main block".
* Residual block adds a Convolutional layer and reduces the dimension through appropriate strides before splitting the input into Main and shortcut paths.
* Residual path helps in avoiding the problems faced because of vanishing gradients, as it carries weights from previous block to the next block.
* The final model is created with a combination of several main and residual blocks.
* The model also uses custom loss function.

````python
def add_main_block(num_times_to_add_block: int, x, x_shortcut, filters: int, kernel_size: tuple, strides: tuple, block_number=None, padding: str ='same', alpha: str = 0.1):
    
    for _ in range(num_times_to_add_block):
        x = Conv2D(filters, kernel_size, strides=strides, padding=padding)(x)
        x = BatchNormalization()(x)
        x = LeakyReLU(alpha=alpha)(x)
    
        x = Conv2D(filters*2, kernel_size, strides=strides, padding=padding)(x)
        x = BatchNormalization()(x)
        x = LeakyReLU(alpha=alpha)(x)
    
        x = Add()([x_shortcut, x])
        x = LeakyReLU(alpha=alpha)(x)
        
        x_shortcut = x
        
    return x, x_shortcut



def add_res_block(x, x_shortcut=None, filters: int = None, kernel_size: tuple = None, strides: tuple = None, block_number=None, padding: str = 'same', alpha: float = 0.1):
    x = tf.keras.layers.Conv2D(filters, kernel_size, strides=strides, padding=padding)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.LeakyReLU(alpha=alpha)(x)

    x_shortcut = x
    
    return x, x_shortcut
````

# Loss Functions

`NOTE:` The TPU does not have a SUM_OVER_BATCH_SIZE function, so the confidence part of the loss function is modified accordingly.
> Binary_Crossentropy loss is used for confidence level

> Mean Square Error (MSE) loss is used for (x-center,y-center) and (w,h)

- The output format of the model is [batch_size, 32, 32, 10]
- Taking [:, :, :, 0:10] for caculating the loss
- The 10 outputs of each grid are formatted as below
- [0]: Confidence probability of anchor one

- [1,2]: x,y-center of anchor one

- [3,4]: w,h of anchor one

- [5]: Confidence probability of anchor two

- [6,7]: x,y-center of anchor two

- [8,9]: w,h-center of anchor two



````python
def custom_loss(y_true, y_pred):
    
    
    
    global BATCH_SIZE
    
    #------------- For GPU Computing ----------------#
    binary_crossentropy = prob_loss = tf.keras.losses.BinaryCrossentropy(
        reduction=tf.keras.losses.Reduction.SUM_OVER_BATCH_SIZE
    )

    prob_loss = binary_crossentropy(
        tf.concat([y_true[:,:,:,0], y_true[:,:,:,5]], axis=0), 
        tf.concat([y_pred[:,:,:,0], y_pred[:,:,:,5]], axis=0)
     )
    #-------------- End of GPU Computing ---------------#
    
    

    #------------ For TPU Computing ------------#
#     binary_crossentropy = prob_loss = tf.keras.losses.BinaryCrossentropy(
#         reduction=tf.keras.losses.Reduction.NONE)
    
#     # Binary Cross Entropy loss for TPU Computing.
#     reduce_sum = tf.reduce_sum(binary_crossentropy(
#         tf.concat([y_true[:,:,:,0], y_true[:,:,:,5]], axis=0),
#         tf.concat([y_pred[:,:,:,0], y_pred[:,:,:,5]], axis=0)
#     ))
    
#     prob_loss = reduce_sum * (1. / BATCH_SIZE)
    #-------------- End of TPU Computing --------------#
    
    xy_loss = tf.keras.losses.MSE(
        tf.concat([y_true[:,:,:,1:3], y_true[:,:,:,6:8]], axis=0), 
        tf.concat([y_pred[:,:,:,1:3], y_pred[:,:,:,6:8]], axis=0)
    )
    
    wh_loss = tf.keras.losses.MSE(
        tf.concat([y_true[:,:,:,3:5], y_true[:,:,:,8:10]], axis=0), 
        tf.concat([y_pred[:,:,:,3:5], y_pred[:,:,:,8:10]], axis=0)
    )
    
    bboxes_mask = get_mask(y_true)
    
    xy_loss = xy_loss * bboxes_mask
    wh_loss = wh_loss * bboxes_mask
    
    return prob_loss + xy_loss + wh_loss


def get_mask(y_true):
    anchor_one_mask = tf.where(
        y_true[:,:,:,0] == 0, 
        0.5, 
        5.0
    )
    
    anchor_two_mask = tf.where(
        y_true[:,:,:,5] == 0, 
        0.5, 
        5.0
    )
    
    bboxes_mask = tf.concat(
        [anchor_one_mask,anchor_two_mask],
        axis=0
    )
    
    return bboxes_mask
````

# Model Construction

````python
#with tpu_strategy.scope(): # 

x_input = Input(shape=(256, 256, 3))

x = Conv2D(32, (3, 3), strides=(1, 1), padding='same')(x_input)
x = BatchNormalization()(x)
x = LeakyReLU(alpha=0.1)(x)


#------------ Block 1 -------------#
#-- Res-block-1 --#
x, x_shortcut = add_res_block(x, filters=64, kernel_size=(3, 3), strides=(1, 1), padding='same', alpha=0.1)
#-- Main-Block-1 --#
x, x_shortcut = add_main_block(2, x, x_shortcut, 32, kernel_size=(3, 3), strides=(1, 1), padding='same', alpha=0.1)


#------------ Block 2 -------------#
#-- Res-block-2 --#
x, x_shortcut = add_res_block(x, x_shortcut=x_shortcut, filters=128, kernel_size=(3, 3), strides=(1, 1), padding='same', alpha=0.1)
#-- Main-Block 2 --#
x, x_shortcut = add_main_block(2, x, x_shortcut, 64, kernel_size=(3, 3), strides=(1, 1), padding='same', alpha=0.1)


#------------ Block 3 -------------#
#-- Res-block 3 --#
x, x_shortcut = add_res_block(x, x_shortcut, filters=256, kernel_size=(3, 3), strides=(2, 2), padding='same', alpha=0.1)
#-- Main-Block --#
x, x_shortcut = add_main_block(8, x, x_shortcut, 128, kernel_size=(3, 3), strides=(1, 1), padding='same', alpha=0.1)


#------------ Block 4 -------------#
#-- Res-Block 4 --#
x, x_shortcut = add_res_block(x, x_shortcut, filters=512, kernel_size=(3, 3), strides=(2, 2), padding='same', alpha=0.1)
#-- Main-Block 4 --#
x, x_shortcut = add_main_block(8, x, x_shortcut, 256, kernel_size=(3, 3), strides=(1, 1), padding='same', alpha=0.1)


#------------ Block 5 -------------#
#-- Res-Block 5 --#
x, x_shortcut = add_res_block(x, x_shortcut, filters=1024, kernel_size=(3, 3), strides=(2, 2), padding='same', alpha=0.1)
#-- Main-Block 5 --#
x, x_shortcut = add_main_block(8, x, x_shortcut, 512, kernel_size=(3, 3), strides=(1, 1), padding='same', alpha=0.1)


#------------ Output Layers -------------#
x = Conv2D(512, (3, 3), strides=(1, 1), padding='same')(x)
x = BatchNormalization()(x)
x = LeakyReLU(alpha=0.1)(x)

x = Conv2D(256, (3, 3), strides=(1, 1), padding='same')(x)
x = BatchNormalization()(x)
x = LeakyReLU(alpha=0.1)(x)

x = Conv2D(128, (3, 3), strides=(1, 1), padding='same')(x)
x = BatchNormalization()(x)
x = LeakyReLU(alpha=0.1)(x)

predictions= x = Conv2D(10, (1, 1), strides=(1, 1), activation='sigmoid', dtype="float32")(x)


model = tf.keras.Model(inputs=x_input, outputs=predictions)
````

````python
optimiser = tf.keras.optimizers.Adam(learning_rate=0.0001)

model.compile(
    optimizer=optimiser, 
    loss=custom_loss
)
````

# Prediction Functions

````python
def prediction_boxes(bboxes, image_grid):
    
    '''
    
    '''
    
    bboxes = bboxes.copy()
    im_width = (image_grid[:, :, 2]*32)
    im_height = (image_grid[:, :, 3]*32)
    
    # descale x,y
    bboxes[:, :, 1] = (bboxes[:, :, 1] * image_grid[:, :, 2]) + image_grid[:, :, 0]
    bboxes[:, :, 2] = (bboxes[:, :, 2] * image_grid[:, :, 3]) + image_grid[:, :, 1]
    bboxes[:, :, 6] = (bboxes[:, :, 6] * image_grid[:, :, 2]) + image_grid[:, :, 0]
    bboxes[:, :, 7] = (bboxes[:, :, 7] * image_grid[:, :, 3]) + image_grid[:, :, 1]
    
    # descale width, height
    bboxes[:, :, 3] = bboxes[:, :, 3] * im_width
    bboxes[:, :, 4] = bboxes[:, :, 4] * im_height
    bboxes[:, :, 8] = bboxes[:, :, 8] * im_width
    bboxes[:, :, 9] = bboxes[:, :, 9] * im_height
    
    # center x,y to top left x,y
    bboxes[:, :, 1] = bboxes[:, :, 1] - (bboxes[:, :, 3]/2)
    bboxes[:, :, 2] = bboxes[:, :, 2] - (bboxes[:, :, 4]/2)
    bboxes[:, :, 6] = bboxes[:, :, 6] - (bboxes[:, :, 8]/2)
    bboxes[:, :, 7] = bboxes[:, :, 7] - (bboxes[:, :, 9]/2)
    
    # width, height to x_max, y_max
    bboxes[:, :, 3] = bboxes[:, :, 1] + bboxes[:, :, 3]
    bboxes[:, :, 4] = bboxes[:, :, 2] + bboxes[:, :, 4]
    bboxes[:, :, 8] = bboxes[:, :, 6] + bboxes[:, :, 8]
    bboxes[:, :, 9] = bboxes[:, :, 7] + bboxes[:, :, 9]
    
    return bboxes
    
````

````python
def non_max_supression(predictions, top_n):
    probabilities = np.concatenate((predictions[:, :, 0].flatten(), predictions[:, :, 5].flatten()), axis=None)
    
    anchor_1 = predictions[:, :, 1:5].reshape((32*32, 4))
    anchor_2 = predictions[:, :, 6:10].reshape((32*32, 4))
    
    bboxes = np.concatenate((anchor_1, anchor_2))
    
    bboxes, probabilities = select_top_possibilities(probabilities, bboxes, top_n=top_n)
    bboxes = switch_xy(bboxes)
    
    return bboxes, probabilities

def switch_xy(bboxes):
    
    tmp_bboxes = np.zeros(np.shape(bboxes))
    
    tmp_bboxes[:, 0], tmp_bboxes[:, 1], tmp_bboxes[:, 2], tmp_bboxes[:, 3] = bboxes[:, 1].copy(), bboxes[:, 0].copy(), bboxes[:, 3].copy(), bboxes[:, 2].copy()
    
    return tmp_bboxes

def select_top_possibilities(probabilities, bboxes, top_n=10):
    top_indicies = tf.image.non_max_suppression(
        boxes=bboxes,
        scores=probabilities,
        max_output_size=top_n,
        iou_threshold=0.3,
        score_threshold=0.3
    )
    
    top_indicies = top_indicies.numpy()
    
    return bboxes[top_indicies], probabilities[top_indicies]
````

````python
# Wrapper for all the post processing functions
# We could check for leading bounding boxes which are out of the image, but not implimented due to time constraint.

def post_process_predictions(predictions, image_ids, image_grid):
    bboxes = {}
    probabilities = {}
    for i, image_id in enumerate(image_ids):
        predictions[i] = prediction_boxes(predictions[i], image_grid)
        bboxes[image_id], probabilities[image_id] = non_max_supression(predictions[i], top_n=100)
    
    # converting to coco points
    
        bboxes[image_id][:, 2:4] = bboxes[image_id][:, 2:4] - bboxes[image_id][:, 0:2]
    
    return bboxes, probabilities
````
